```python
df["viral"] = (df["trend_label"] == "rising").astype(int)
```

```python
df.groupby("viral")[["completion_rate", "save_rate", "share_rate",
                     "avg_watch_time_sec", "duration_sec"]].mean()
```

```python
#result
df.groupby(df["engagement_rate"] >= df["engagement_rate"].median())[["save_rate","share_rate","comment_ratio","duration_sec"]].mean()
```
