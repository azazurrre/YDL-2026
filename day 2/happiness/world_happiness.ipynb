```python
#Богатые, но не самые счастливые
df[(df["GDP per capita"] > 1.3) & (df["Score"] < 6)]
```

```python
#Бедные, но относительно счастливые
df[(df["GDP per capita"] < 0.7) & (df["Score"] > 5)]
```

```python
# Страны, где счастье держится на свободе, а не на деньгах:
# свобода выше средней, ВВП ниже среднего, при этом страна достаточно счастлива
m = df.describe().loc["mean"]

freedom_driven = df[
    (df["Freedom to make life choices"] > m["Freedom to make life choices"]) &
    (df["GDP per capita"] < m["GDP per capita"]) &
    (df["Score"] > 5)
].sort_values("Freedom to make life choices", ascending=False)

freedom_driven[["Overall rank", "Country or region", "Score",
                "GDP per capita", "Freedom to make life choices"]]
```

```python
#Страны, которые счастливее, чем «положено»
factors = ["GDP per capita","Social support","Healthy life expectancy",
           "Freedom to make life choices","Generosity","Perceptions of corruption"]
df["factors_sum"] = df[factors].sum(axis=1)
df["residual"] = df["Score"] - df["factors_sum"]
df.sort_values("residual", ascending=False).head(10)
```

```python
#Страны, которые несчастнее, чем «положено»
factors = ["GDP per capita","Social support","Healthy life expectancy",
           "Freedom to make life choices","Generosity","Perceptions of corruption"]
df["factors_sum"] = df[factors].sum(axis=1)
df["residual"] = df["Score"] - df["factors_sum"]
df.sort_values("residual").head(10)
```
